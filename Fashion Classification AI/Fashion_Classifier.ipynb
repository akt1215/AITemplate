```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.datasets import fashion_mnist
```

```python
#importing data
data = fashion_mnist.load_data()
class_name = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt','Sneaker', 'Bag', 'Ankle boot']
```

```python
data
```

```python
#splitting data into train and test
(X_train, y_train), (X_test, y_test) = data
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
X_train
```

```python
#implementing model
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

#compiling model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

#fitting model
number_of_epochs = 20
history = model.fit(X_train, y_train, epochs=number_of_epochs, verbose=1, validation_split=0.2)

#evaluating model
model.evaluate(X_test, y_test)

#showing the accuracy
print("Accuracy: ", model.evaluate(X_test, y_test)[1]*100, "%")

#plotting the accuracy and number of epochs
plt.scatter(range(number_of_epochs), history.history['accuracy'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()

#Showing the image and the prediction
for i in range(5):
    plt.imshow(X_test[i], cmap='gray', interpolation='none')
    plt.xlabel(class_name[y_test[i]])
    plt.ylabel(class_name[np.argmax(model.predict(X_test[i].reshape(1,28,28)))])
    plt.show()

model.save("fashion_mnist.h5")
```
